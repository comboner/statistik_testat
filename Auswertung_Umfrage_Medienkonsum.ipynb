## Auswertung Umfrage zu Medienkonsum

print data

```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import bqplot as bq

# Read response data
data = pd.read_csv('data/UMFRAGE_after_clean_up.csv')
del data['gender']
del data['Timestamp']
del data['update per week']
del data['Unnamed: 26']
del data['information_sources_used']
del data['situations']


pd.set_option('display.max_columns', None) 
print(data.head())
```

# statistical fingerprint

get statistical fingerprint

```python
print('Data Types:\n' + str(data.dtypes))
print('\nStatistical Fingerprint:\n' + str(data.describe()))
##print('\nValue count for Non-Numeric Columns \n' + str(data['gender'].value_counts())) non non-numeric values used
```

print correlations

```python
plt.subplots(figsize=(44, 33))
sns.set(font_scale=1.5)
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn_r', linewidths=0.5)
```
